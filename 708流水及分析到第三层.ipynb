```python
def chuli_chongzheng_lei(i2, f2):
    if pd.isna(i2):
        return f2  # or any appropriate value when i2 is NA

    if i2 in ["出", "D", "借", "转出"]:
        result = -f2
    else:
        result = f2

    return result
ls_qc.insert(4,'流向金额',ls_qc.apply(lambda row: chuli_chongzheng_lei(row['收付标志'], row['交易金额']), axis=1))
```

```python
import pandas as pd
from datetime import datetime, timedelta
criteria_df=pd.read_excel(r"D:\项目\西宁\分析\710调单过程.xlsx",sheet_name='20万以下的小于5天',dtype={'卡号':str})
transactions_df=ls_qc
transactions_df.loc[:, '交易时间'] = pd.to_datetime(transactions_df['交易时间'])
criteria_df['最小时间'] = pd.to_datetime(criteria_df['最小时间'])
criteria_df['最大时间'] = pd.to_datetime(criteria_df['最大时间'])

# 定义前后3分钟的时间间隔
delta = timedelta(minutes=3)

# 使用apply函数进行筛选
filtered_df = transactions_df.merge(criteria_df, on='卡号', how='left').\
    loc[lambda x: (x['交易时间'] > x['最小时间'] - delta) & (x['交易时间'] < x['最大时间'] + delta)]
```

```python
filtered_df['流向金额'] = pd.to_numeric(filtered_df['流向金额'],errors='coerce')
```

```python
sj_tfw=filtered_df.groupby(['交易对手账卡号', '对手户名']).agg(sum=('流向金额', 'sum'), count=('流向金额', 'count')).reset_index().sort_values('sum', ascending=True)
```
