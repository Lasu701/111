```python

import os
import re
from tqdm import tqdm
import csv
import pandas as pd
import json

tksq=[]

with open(r"D:\项目\西宁\取证\取证数据\财务管理-会员填单管理-会员填单存款列表-z-20230526.json","r",encoding='utf-8') as f:
    data = f.read()
    p1 = re.compile(r'(\[\{.*?\}\])', re.S)
    kk=re.findall(p1, data)
    for  i in  range(len(kk)):
        tmp = json.loads(kk[i])
        try:
            #df = pd.read_json(kk,encoding="utf-8", orient='records')
            df_data = pd.DataFrame.from_dict(tmp)
            # df = json_normalize(data)

        except Exception as e:

        #print(kk)
            continue
            # df = pd.read_json(k

        tksq.append(df_data)
td_ck=pd.concat(tksq)

```
