```python
def chuli_chongzheng_lei(i2, f2):
    if pd.isna(i2):
        return f2  # or any appropriate value when i2 is NA

    if i2 in ["出", "D", "借", "转出"]:
        result = -f2
    else:
        result = f2

    return result
data_sf.insert(4,'流向金额',data_sf.apply(lambda row: chuli_chongzheng_lei(row['收付标志'], row['交易金额']), axis=1))
```
